###Regresión Lineal Multiple

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
shoedata = pd.read_csv("shoes_dataset.csv")
shoedata
```

Ahora vamos a hacer un ejemplo de "regresión multiple", con los datos de nuestra base de datos. Primero empezaremos creando un dataframe con las columnas de "sale", "retail", y "profit". 

```python
from statsmodels.formula.api import ols

Sale = shoedata["Sale"]
Retail  = shoedata ["Retail"]
Profit = shoedata ["Profit"]
df1 = pd.DataFrame( {"Sale": Sale, "Retail": Retail, "Profit": Profit})
print(df1)
```

Aqui abajo tuvimos un contratiempo con la columna de VENTAS, ya que no nos dejaba analizarla por las comas (,) que había, es por eso que las quitamos en el código siguiente. 

```python
df1['Sale']= df['Sale'].apply(str).str.replace(',','').astype(float)
```

```python
df1
```

Aqui abajo, lo que queremos lograr con el código, es la:

REGRESIÓN respecto a... (la variable responsiva: que en este caso es "Sale") ~ (las otras dos variables: "Retail" y "Profit")

Podemos ver el valor en el (R-squared), es "1", por lo que concluimos que es un muy buen modelo. 

```python
RM = ols(formula = "Sale ~ Retail + Profit", data = df1 )
Fit = RM.fit() #(aqui alimentamos al modelo)
Fit.summary()
```

Esta "tabla", que vemos aqui arriba, que muestra la (regresión multiple):

Nos muestra la siguiente información: 

Y(Sale) = B0(intercept), X1(Retail), X2(Profit), X3(Retail * Profit)
